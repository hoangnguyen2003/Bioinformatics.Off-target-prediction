```python
import os

# Specify the current and new folder names
current_folder_name = 'Bioinformatics.Off-target-prediction'
new_folder_name = 'bbbbb'

# Define the path to the output directory
output_directory = '/kaggle/working/'

# Create the full paths for the current and new folders
current_folder_path = os.path.join(output_directory, current_folder_name)
new_folder_path = os.path.join(output_directory, new_folder_name)

# Rename the folder
try:
    os.rename(current_folder_path, new_folder_path)
    print(f"Folder '{current_folder_name}' renamed to '{new_folder_name}' successfully.")
except FileNotFoundError:
    print(f"Folder '{current_folder_name}' not found.")
except FileExistsError:
    print(f"A folder named '{new_folder_name}' already exists.")
except Exception as e:
    print(f"An error occurred: {e}")
```
